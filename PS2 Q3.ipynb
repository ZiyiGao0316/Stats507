```python
import pandas as pd
import numpy as np

## a
def gen_cohort_category(year1, year2):
    '''
    input:
        year1: int
        year2: int
    output:
        string
    '''
    string_format = "{}-{}"
    
    return string_format.format(year1, year2)

## a
data17_18 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DEMO_J.XPT')
data15_16 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.XPT')
data13_14 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT')
data11_12 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DEMO_G.XPT')

column_list = ["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", "RIDSTATR",
              "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR"]

# create a new column and combine the data

data11_12 = data11_12[column_list]
data13_14 = data13_14[column_list]
data15_16 = data15_16[column_list]
data17_18 = data17_18[column_list]

data11_12["cohort"] = gen_cohort_category(11, 12)
data13_14["cohort"] = gen_cohort_category(13, 14)
data15_16["cohort"] = gen_cohort_category(15, 16)
data17_18["cohort"] = gen_cohort_category(17, 18)

dataframe1 = [data11_12, data13_14, data15_16, data17_18]
data1 = pd.concat(dataframe1)

# rename the columns
new_column_list = ['unique id', 'age', 'race and ethnicity', 'education', 
                   'marital status', 'interview and examination status of the participant',
                  'masked variance unit pseudo-psu variable for variance estimation', 
                  'masked variance unit pseudo-stratum variable for variance estimation',
                  'full sample 2 year mec exam weight', 'full sample 2 year interview weight', 
                  'cohort']
data1.columns = new_column_list
# convert the variables to appropriate types
data1 = data1.convert_dtypes()

data1.to_pickle("./data1.pkl")
new_data1 = pd.read_pickle("./data1.pkl")

## b

var1_format = "OHX{}TC"
var2_format = "OHX{}CTC"
var1_range = list(range(1, 33))
var2_range = list(range(2, 32))

for i in range(len(var1_range)):
    if int(var1_range[i]) < 10:
        var1_range[i] = "%02d" % (int(var1_range[i]),)
var1 = [var1_format.format(i) for i in var1_range]

for j in range(len(var2_range)):
    if int(var2_range[j]) < 10:
        var2_range[j] = "%02d" % (int(var2_range[j]),)
var2 = [var2_format.format(j) for j in var2_range]
var2.remove(var2_format.format(16))
var2.remove(var2_format.format(17))
var = ["SEQN", "OHDDESTS"] 

den_1718 = pd.read_sas("https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/OHXDEN_J.XPT")
den_1516 = pd.read_sas("https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/OHXDEN_I.XPT")
den_1314 = pd.read_sas("https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/OHXDEN_H.XPT")
den_1112 = pd.read_sas("https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/OHXDEN_G.XPT")

column_lists = var + var1 + var2
den_1112 = den_1112[column_lists]
den_1314 = den_1314[column_lists]
den_1516 = den_1516[column_lists]
den_1718 = den_1718[column_lists]

den_1112["cohort"] = gen_cohort_category(11, 12)
den_1314["cohort"] = gen_cohort_category(13, 14)
den_1516["cohort"] = gen_cohort_category(15, 16)
den_1718["cohort"] = gen_cohort_category(17, 18)

dataframe2 = [den_1112, den_1314, den_1516, den_1718]
data2 = pd.concat(dataframe2)

new_var_end = ['cohort']
new_var1_range = list(range(1, 33))
new_var2_range = list(range(2, 32))

new_var1_format = "tooth count: #{}"
new_var2_format = "coronal caries: tooth count #{}"

new_var1 = [new_var1_format.format(m) for m in new_var1_range]
new_var2 = [new_var2_format.format(n) for n in new_var2_range]
new_var = ["unique id", "dentation status code"]
new_var2.remove(new_var2_format.format(16))
new_var2.remove(new_var2_format.format(17))

new_columns_lists = new_var + new_var1 + new_var2 + new_var_end

data2.columns = new_columns_lists
data2 = data2.convert_dtypes()

data2.to_pickle("./data2.pkl")
new_data2 = pd.read_pickle("./data2.pkl")

## c
data1_cases = data1.shape[0]
data2_cases = data2.shape[0]
```
